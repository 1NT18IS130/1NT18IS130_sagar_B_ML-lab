```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
```

```python
col_names = ["animal name","hair","feathers","eggs","milk","airborne","aquatic","predator","toothed","backbone","breathes","venomous","fins","legs","tail","domestic","catsize","typ"]
dataset = pd.read_csv("./zoo.data", header=None, names=col_names)
dataset.head()
```

```python
X, y = dataset.iloc[:,1:17], dataset.iloc[:,17]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=99)
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
plt.figure(figsize=(15,15))
tree.plot_tree(clf)
plt.show()
```

```python
y_pred = clf.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
