# K-Means Clustering

```python
import numpy as np
import random as rd
import math
import matplotlib.pyplot as plt
```

```python
x = [[5.9, 3.2], [4.6, 2.9], [6.2, 2.8], [4.7, 3.2], [5.5, 4.2], [5.0, 3.0], [4.9, 3.1], [6.7, 3.1], [5.1, 3.8], [6.0, 3.0]]
```

```python
def euclidean_distance(p1, p2):
    dist = math.sqrt((p1[0]-p2[0])**2+(p1[1]-p2[1])**2)
    return dist

def minimum(one, two, three):
    if one < two and one < three:
        return 1
    if two < one and two < three:
        return 2
    if three < one and three < two:
        return 3
    else:
        return 4
```

```python
centroid = [[6.2, 3.2], [6.6, 3.7], [6.5, 3.0]]
global cluster1 # red
global cluster2 # green
global cluster3 # blue

cluster1= [] # red
cluster2= [] # green
cluster3= [] # blue

def make_clusters(arr):
    for each in arr:
        one = euclidean_distance(centroid[0], each)
        two = euclidean_distance(centroid[1], each)
        three = euclidean_distance(centroid[2], each)
        what = minimum(one, two, three)
        if what == 1:
            cluster1.append(each)
        if what == 2:
            cluster2.append(each)
        if what == 3:
            cluster3.append(each)
        if what == 4:
            print("something")

make_clusters(x)
print(cluster1)
print(cluster2)
print(cluster3)
```

```python
def get_centroid(cluster):
    x_sum = 0
    y_sum = 0
    
    for each in cluster:
        x_sum += each[0]
        y_sum += each[1]
    centr = [round(x_sum/len(cluster),3), round(y_sum/len(cluster),3)]
    return centr

def flush_clusters():
    cluster1 = []
    cluster2 = []
    cluster3 = []
```

```python
# Answer for centroid of red cluster after one interation
centroid[0] = get_centroid(cluster1)
print("Centroid of Cluster 1 on One iteration: ",centroid[0])
```

```python
# green cluster centroid after two iterations
cluster2_copy = cluster2
cent = get_centroid(cluster2_copy)
centroid[1] = cent
flush_clusters()
make_clusters(x)
cent = get_centroid(cluster2)
print("Centroid of Cluster 2 on second iteration: ",cent)
```

```python
for i in range(0,5):
    cent = get_centroid(cluster3)
    centroid[2] = cent
    print(f"Iteration {i+1}: {cent}")
    flush_clusters()
    make_clusters(x)
```

The centroid of third cluster after converging is [6.22, 3.01]

```python
for i in range(0,500):
    if i == 0:
        make_clusters(x)
    centroid[0] = get_centroid(cluster1)
    centroid[1] = get_centroid(cluster2)
    centroid[2] = get_centroid(cluster3)
    
    print(f"{i+1} Iteration: ", centroid)
    flush_clusters()
    make_clusters(x)
```

The clusters take 400 iterations to converge
